# Token 추가하기

```python
from kobert_transformers import get_kobert_model, get_tokenizer
# Load KoBERT model and tokenizer
model = get_kobert_model()
tokenizer = get_tokenizer()


# Define new tokens
new_tokens = ["새로운토큰1", "새로운토큰2", "가나다라", "남윤지", "햄찌뽕짝"]

# Update vocab and tokenizer
# Ensure no duplicate tokens
existing_tokens = set(tokenizer.idx2token)
new_tokens = [token for token in new_tokens if token not in existing_tokens]

tokenizer.idx2token.extend(new_tokens)
for idx, token in enumerate(new_tokens, start=len(tokenizer.token2idx)):
    tokenizer.token2idx[token] = idx

model.resize_token_embeddings(len(tokenizer.idx2token))

print(f"Number of tokens added: {len(new_tokens)}")
print(f"New vocabulary size: {len(tokenizer.idx2token)}")
```

# 시 한 편 넣어봄

```python
new_poem = "내고장 칠월은 청포도가 익어 가는 시절 이 마음 전설이 주저리주저리 열리고 먼데 하늘이 꿈꾸며 알알이 들어와 박혀 하늘 밑 푸른 바닷가 가슴을 열고 靑袍(청포)를 입고 찾아온다 했으니 내 그를 맞아 이 포도를 따먹으며 두 손 함뿍 적셔도 좋으련 아이야 우리 식탁엔 은쟁반에 하이얀 모시 수건을 마련해 두렴"

# 굉장히 길다.
```

```python
import re

tokens_from_poem = set(re.findall(r'\b\w+\b', new_poem))

# Ensure no duplicate tokens by filtering out existing ones
existing_tokens = set(tokenizer.idx2token)
new_tokens = [token for token in tokens_from_poem if token not in existing_tokens]

# Add new tokens to tokenizer's vocabulary
tokenizer.idx2token.extend(new_tokens)
for idx, token in enumerate(new_tokens, start=len(tokenizer.token2idx)):
    tokenizer.token2idx[token] = idx

# Resize model's embedding layer
model.resize_token_embeddings(len(tokenizer.idx2token))

# Print the number of tokens added and the new vocabulary size
print(f"Number of tokens added: {len(new_tokens)}")
print(f"New vocabulary size: {len(tokenizer.idx2token)}")
```

```python
import torch
import torch.nn.functional as F

def cosine_similarity(tensor1, tensor2):
    # Ensure both tensors have the same shape
    assert tensor1.shape == tensor2.shape, "Tensors must have the same shape for cosine similarity."

    # Calculate cosine similarity
    similarity = F.cosine_similarity(tensor1, tensor2, dim=0)
    return similarity.item()  # Convert to a scalar value

# Example usage:
# Assuming tensor1 and tensor2 are the two tensors you want to compare

similarity_score = cosine_similarity(cls_embedding1, cls_embedding2)
print("Cosine Similarity:", similarity_score)
```

```python
from torch.utils.data import Dataset, DataLoader

class PoemDataset(Dataset):
  def __init__(self, texts, tokenizer, max_length=512):
    self.texts = texts
    self.tokenizer = tokenizer
    self.max_length = max_length

  def __len__(self):
    return len(self.texts)

  def __getitem__(self, idx):
    text = self.texts[idx]
    encoding = self.tokenizer(text, padding='max_length',
                              truncation = True,
                              max_length = self.max_length,
                              return_tensors = 'pt')
    item = {key: val.squeeze() for key, val in encoding.items()}
    return item



texts = df['origin_text_list'].tolist()
dataset = PoemDataset(texts, tokenizer)

# DataLoader 생성
dataloader = DataLoader(dataset, batch_size=8, shuffle=True)
```

```python
import torch
import torch.nn as nn
from transformers import BertTokenizer, BertModel
from torch.utils.data import Dataset, DataLoader
import random
from tqdm import tqdm

# Check if CUDA is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# KoBERT 모델과 토크나이저 로드
model = BertModel.from_pretrained('monologg/kobert').to(device)
tokenizer = BertTokenizer.from_pretrained('monologg/kobert')

# 모든 파라미터 동결
for param in model.parameters():
    param.requires_grad = False

# Adapter 레이어 정의
class Adapter(nn.Module):
    def __init__(self, input_dim, hidden_dim):
        super(Adapter, self).__init__()
        self.down_project = nn.Linear(input_dim, hidden_dim)
        self.up_project = nn.Linear(hidden_dim, input_dim)
        self.activation = nn.ReLU()

    def forward(self, x):
        x = self.down_project(x)
        x = self.activation(x)
        return self.up_project(x)

# Adapter 레이어 초기화
adapter_layer = Adapter(input_dim=768, hidden_dim=64).to(device)

# 데이터셋 정의
class LiteraryMLMDataset(Dataset):
    def __init__(self, texts, tokenizer, max_length=512, mask_prob=0.15):
        self.texts = texts
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.mask_prob = mask_prob

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        encoding = self.tokenizer(text, padding='max_length', truncation=True, max_length=self.max_length, return_tensors="pt")
        input_ids = encoding['input_ids'].squeeze()

        # MLM: 마스킹 생성
        labels = input_ids.clone()
        rand = torch.rand(input_ids.shape)
        mask_arr = (rand < self.mask_prob) * (input_ids != self.tokenizer.cls_token_id) * \
                   (input_ids != self.tokenizer.sep_token_id) * (input_ids != self.tokenizer.pad_token_id)

        # 마스킹된 위치를 [MASK] 토큰으로 대체
        input_ids[mask_arr] = self.tokenizer.mask_token_id

        # 마스크된 위치의 정답은 그대로 남기고, 나머지는 -100으로 설정 (loss 계산 제외)
        labels[~mask_arr] = -100
        return {
            'input_ids': input_ids.to(device),
            'attention_mask': encoding['attention_mask'].squeeze().to(device),
            'labels': labels.to(device)
        }

# DataLoader 생성
texts = df['origin_text_list'].tolist()
dataset = LiteraryMLMDataset(texts, tokenizer)
dataloader = DataLoader(dataset, batch_size=8, shuffle=True)

# 옵티마이저 설정 (adapter 레이어만 학습)
optimizer = torch.optim.AdamW(adapter_layer.parameters(), lr=1e-4, weight_decay = 0.1)

# 학습 루프
model.train()
adapter_layer.train()
for epoch in range(3):
    total_loss = 0
    for batch in tqdm(dataloader, desc=f"Epoch {epoch+1} Progress"):
        input_ids = batch['input_ids']
        attention_mask = batch['attention_mask']
        labels = batch['labels']

        # KoBERT 모델에서 임베딩 추출
        with torch.no_grad():  # KoBERT의 파라미터는 업데이트하지 않음
            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            sequence_output = outputs.last_hidden_state  # [batch_size, sequence_length, hidden_size]

        # Adapter 레이어 통과
        adapter_output = adapter_layer(sequence_output)

        # Masked Language Modeling을 위한 예측을 위해 Linear 레이어 추가
        # 어휘 크기에 맞게 출력 크기를 조정해야 함
        vocab_size = tokenizer.vocab_size
        prediction_layer = nn.Linear(adapter_output.size(-1), vocab_size).to(device)
        predictions = prediction_layer(adapter_output)

        # 손실 계산
        loss_fn = nn.CrossEntropyLoss(ignore_index=-100)  # 마스크된 토큰에 대해서만 계산
        loss = loss_fn(predictions.view(-1, vocab_size), labels.view(-1))

        # 역전파와 옵티마이저 단계
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        total_loss += loss.item()
    print(f"Epoch {epoch + 1}, Loss: {total_loss / len(dataloader)}")
```
